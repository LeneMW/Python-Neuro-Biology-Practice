How can I plot with floats? Let's try!

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

standard_curve = np.array([0.0, 0.1, 0.2, 0.3])
AU = np.array([0.5, 0.7, 0.9, 1.2])

## The plot above did not use realistic standard curve values and AU to what I typically get, but I got an error for trying to plot with a float! Now I made a numpy array and was able to input more realistic values! Exciting.

plt.plot(standard_curve, AU)
plt.xlabel("Protein Concentration ug/ul")
plt.ylabel("Arbitrary Units")
m, b = np.polyfit(standard_curve, AU, 1) ## This is to get the linear equation.
plt.plot(standard_curve, m*standard_curve + b, marker='o', linestyle='-', color='b') ## This is to plot and style the graph
slope, intercept, r_value, _, _ = linregress(standard_curve, AU) 
r_squared = r_value ** 2 ## This is to get the R-squared value
plt.show()
print(f"Linear Equation: y = {m:.2f}x + {b:.2f}")
print(f"R-squared Value: {r_squared:.4f}")

```

The linear equation and R-squared value of the protein concentration standard curve are really important to know for the experiments I am running currently. For example, I am running a lot of western blots and I need to know the protein concentration of each of my samples in order to calculate the volume I need. The reason I need to know the R-squared value is because you want an R-squared value higher than at least 0.95, but ideally you should be in the 0.97 to as close to 1.0 as you can get. The R-squared value tells you how accurate your equation is so you can calculate the protein concentrations from the arbitrary units picked up in the plate reader on your protein samples.

Essentially: R-squared is defined as a number that tells you how well the independent variable(s) in a statistical model explains the variation in the dependent variable. It ranges from 0 to 1, where 1 indicates a perfect fit of the model to the data.
